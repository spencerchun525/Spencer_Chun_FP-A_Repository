```python
numbers = ['$10,23', '$7,15', '$84,83', '$101,91']
def clean(s):
    """Takes a string as an inout and removes dollar signs and replaces commas with periods"""
    cleaned_string = "" #return string starts as empty
    for i in s:
        if i == "$":
            dollar_replacement = i.replace("$","") #removes dollar signs by replacing them with empty string
            cleaned_string += dollar_replacement #adds empty string to return string
        elif i == ",":
            comma_replacement = i.replace(",",".") #replaces commas with periods
            cleaned_string += comma_replacement #adds period to return string
        else:
            cleaned_string += i #If it is neither a dollar nor a comma, just add the letter or number
    return cleaned_string
```

```python
clean("$$$,184,94")
```

```python
cleaned_numbers = [clean(string) for string in numbers]
print(cleaned_numbers)
```

```python
float_numbers = [float(clean(s)) for s in numbers]
print(float_numbers)
```
