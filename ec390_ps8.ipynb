```python
import pandas as pd
import matplotlib.pyplot as plt
banking_data = pd.read_csv("https://users.ssc.wisc.edu/~mckelvey/ec390/banks_and_branches.csv", index_col = 'Year',thousands =',', header = 6)
fig1, ax1 = plt.subplots(3,1, figsize = (15,15))
ax1[0].plot(banking_data.index, banking_data['Branches'], color = 'black', linestyle = 'dotted')
ax1[0].set(ylabel = 'Number of banks', title = 'Number of U.S. commercial banks, 1934-2017')
ax1[1].plot(banking_data.index, banking_data['Offices'], linestyle = 'dashed', color = 'red')
ax1[1].set(ylabel = 'Number of Offices', title = 'Number of offices of U.S. commercial banks, 1934-2017')
ax1[2].plot(banking_data.index, banking_data['Offices']/banking_data['Branches'], color = 'blue', linestyle = '-.')
ax1[2].set(ylabel = 'Number of offices per bank', title = 'Number of offices per bank, 1934-2017')
fig1, ax1 = plt.subplots(figsize = (15,5))
ax1.plot(banking_data.index, banking_data['Branches'], color = 'black', linestyle = 'dotted', label = 'Number of U.S. commercial banks, 1934-2017.')
ax1.plot(banking_data.index, banking_data['Offices'], linestyle = 'dashed', color = 'red', label = 'Number of offices of U.S. commercial banks, 1934-2017.')
ax1.plot(banking_data.index, banking_data['Offices'] / banking_data['Branches'], color = 'blue', linestyle = '-.', label = 'Number of offices per bank, 1934-2017.')
ax1.legend();
```

```python
fig, ax2 = plt.subplots()
earnings_height = pd.read_stata('https://users.ssc.wisc.edu/~mckelvey/ec400/earningsheight.dta')
ax2.hist(earnings_height['age'], bins = 20)
ax2.set(ylabel = 'Frequency', xlabel = 'age')

fig, ax3 = plt.subplots()
earnings_count = pd.crosstab(earnings_height['age'], 'Frequency')
ax3.bar(earnings_count.index, height = earnings_count['Frequency'])
ax3.set(xlabel = 'Age', ylabel = 'Frequency');
```
