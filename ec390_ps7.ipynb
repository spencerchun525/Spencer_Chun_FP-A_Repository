```python
import pandas as pd
import numpy as np
hisp_df = pd.read_excel('https://users.ssc.wisc.edu/~mckelvey/ec390/wi_pop_countylevel_hisp.xlsx', sheet_name = 'Hispanic',header = 3, index_col = 'Unnamed: 0')
non_hisp_df = pd.read_excel('https://users.ssc.wisc.edu/~mckelvey/ec390/wi_pop_countylevel_hisp.xlsx',sheet_name = 'Non-Hispanic', header = 3, index_col = 'Unnamed: 0')
display(hisp_df)
display(non_hisp_df)
population = hisp_df+non_hisp_df
display(population)
proportion_hispanic = hisp_df/population
display(proportion_hispanic)
percentage_hispanic = proportion_hispanic*100
display(percentage_hispanic)
population.sum()
```

```python
university_data = pd.read_excel('https://users.ssc.wisc.edu/~mckelvey/ec390/tabn312.10.xlsx', header = 1, skipfooter = 4, index_col = 'Institution')
display(university_data)
university_data.drop(1, axis = 'rows', inplace = True)
university_data.rename(columns = {'State':'state', 'Rank\\1\\':'rank', 'Control\\2\\':'type', 'Level':'level','Total enrollment':'enrollment_total','Enrollment in exclusively distance education courses':'enrollment_distance'},inplace = True)
university_data['percent_distance'] = 100*(university_data['enrollment_distance'] / university_data['enrollment_total'])
display(university_data)
print("The university with the lowest percentage of students in distance learning programs is {} with a value of {:.1f} percent".format(university_data['percent_distance'].idxmin(), university_data['percent_distance'].min()))
tab = pd.crosstab(university_data['type'],'count')
tab['proportion'] = pd.crosstab(university_data['type'], 'count', normalize = 'columns')
display(tab)
pd.crosstab(university_data['type'], university_data['level'])
```
