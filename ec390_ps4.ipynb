```python
words = ['Are', 'you', 'suggesting', 'coconuts', 'migrate', '?']
sentence = "" #empty string to reference in the loop, will be added onto
for word in words:
    if word == words[-1] or word == words[-2]:  #Attaches the final two entries without a space, giving "migrate?"
        sentence += word
    else:
        sentence += (word + ' ')     #Attaches the rest of the entries with a space after
print(sentence)
```
