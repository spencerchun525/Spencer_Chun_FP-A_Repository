```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pandas_datareader import wb
year = 1995
indicators = {"SE.PRM.CMPT.MA.ZS":"male_rate", "SE.PRM.CMPT.FE.ZS":"female_rate"}
primary_percent_complete = wb.download(indicator = indicators.keys(), country = 'WLD', errors = 'ignore', start = year)
import seaborn as sns
primary_percent_complete.reset_index(inplace=True)
primary_percent_complete['year'] = primary_percent_complete['year'].astype(int)
primary_percent_complete.set_index('year', inplace=True)
primary_percent_complete.rename(columns = {"SE.PRM.CMPT.MA.ZS":"male_rate", "SE.PRM.CMPT.FE.ZS":"female_rate"}, inplace = True)
fig, ax = plt.subplots()
ax.plot(primary_percent_complete['male_rate'], label = "Males")
ax.plot(primary_percent_complete['female_rate'], label = 'Females')
ax.legend()
ax.set(ylabel = 'Percentage of children who complete primary school', title = 'Percentage of children who complete primary school by year')
sns.despine()
```
